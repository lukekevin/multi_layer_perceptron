```python
import numpy as np
```

```python

#activation functions
def sigmoid_activation(weight_sum):
    return 1 / (1 + np.exp(-weight_sum))

def sigmoid_derivative(weight_sum):
    return weight_sum * (1 - weight_sum)


def normal_activation(weight_sum):
    return 1 if weight_sum>0 else 0
```

```python
raw_data=np.loadtxt('stellar_spectra_data/train.dat')
target_raw=np.loadtxt('stellar_spectra_data/target.dat')
input=np.array(raw_data, dtype=float)
target=np.array(target_raw, dtype=float)

```

```python
input_min = input.min(axis=0)
input_max = input.max(axis=0)
input = (input - input_min) / (input_max - input_min)

input = np.c_[input,  np.ones(input.shape[0])]  #with bias
```

```python
input_size = input.shape[1]  
hidden_size = 10         
output_size = 7 


np.random.seed(42)  
weights_input_hidden = np.random.uniform(-0.5, 0.5, (input_size, hidden_size))
weights_hidden_output = np.random.uniform(-0.5, 0.5, (hidden_size, output_size))

learn_rate = 0.1
epochs = 3

#ONE HOT ENCODING (ACTUALLY THE TARGET IS ALREADY ONE HOT ENCODED)
target_one_hot=target
```

```python
#main training loop 
for epoch in range(epochs):
    for i in range(input.shape[0]):

        #Front pass oth layer
        hidden_input = np.dot(input[i], weights_input_hidden)
        hidden_output = sigmoid_activation(hidden_input)
        # 1st layer
        final_input = np.dot(hidden_output, weights_hidden_output)
        final_output = sigmoid_activation(final_input)

        error = target_one_hot[i] - final_output
        print(target_one_hot[i], final_output)

        # Back pass from 1st layer to 0th layer
        grad_output = error * sigmoid_derivative(final_output)
        hidden_error = np.dot(grad_output, weights_hidden_output.T)
        grad_hidden = hidden_error * sigmoid_derivative(hidden_output)

        # Weights Update 
        weights_hidden_output += learn_rate * np.outer(hidden_output, grad_output)  
        weights_input_hidden += learn_rate * np.outer(input[i], grad_hidden)  
        
```

```python
#VALIDATION PHASE
correct_predict = []
for i in range(input.shape[0]):
    hidden_output = sigmoid_activation(np.dot(input[i], weights_input_hidden))
    final_output = sigmoid_activation(np.dot(hidden_output, weights_hidden_output))
    prediction = np.argmax(final_output) 
    if prediction==np.argmax(target[i]): #AS TARGET WAS ONE HOT ENCODED BUT THE PREDICTION WAS NOT SO I  CONVERTED  ONE HOT ENCODE TARGET TO CLASSES BY ARGMAX
        correct_predict.append(prediction)

accuracy= len(correct_predict) / target.shape[0]
print(accuracy*100)
```
